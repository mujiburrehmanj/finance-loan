# TOTAL LOAN APLLICATION

```python
total_loan_application=df['id'].count()
print('TOTAL LOAN APLLICATION :',total_loan_application)
```

```python
#datatype change issue_date
df['issue_date']=pd.to_datetime(df['issue_date'],format="%d-%m-%Y")
```

## MTD_TOTAL LOAN APPLICATION

```python
latest_month=df['issue_date'].max().month
latest_year=df['issue_date'].max().year
mtd_loan_application=df[
    (df['issue_date'].dt.month == latest_month) &
    (df['issue_date'].dt.year == latest_year)
]['id'].count()
print('MTD total loan applications for month:', latest_month,
      'and year:', latest_year,
      '=> total:', mtd_loan_application)
```

### TOTAL FUNDED AMOUNT

```python
total_funded_amount=df['loan_amount'].sum()
print('TOTAL FUNDED AMOUNT :',total_funded_amount)
```

### MTD_TOTAL FUNDED AMOUNT

```python

mtd_funded_amount=df[
    (df['issue_date'].dt.month == latest_month) &
    (df['issue_date'].dt.year == latest_year)
]['loan_amount'].sum()
print("TOTAL FUNDED AMOUNT: ${:.2f}M".format(mtd_funded_amount/1_000_000))
```

### TOTAL AMOUNT RECIVED

```python
total_amount_recived=df['total_payment'].sum()
print("TOTAL AMOUNT RECIVED:${:.2f}M".format(total_amount_recived/1_000_000))
```

```python

MTD_AMOUNT_RECIVED=df[
    (df['issue_date'].dt.month == latest_month) &
    (df['issue_date'].dt.year == latest_year)
]['total_payment'].sum()
print("TOTAL AMOUNT RECIVED : ${:.2f}M".format(MTD_AMOUNT_RECIVED/1_000_000))
```

# GRAPHS

```python
import matplotlib.pyplot as plt
```

```python
#extact year and monthname
df['year']=df['issue_date'].dt.year
df['month']=df['issue_date'].dt.month_name()
#total loan application by year and month
monthly_loans=df.groupby(['year','month'])['id'].count().reset_index()
month_order = ['January','February','March','April','May','June','July','August','September','October','November','December']
monthly_loans['month']=pd.Categorical(monthly_loans['month'],categories=month_order,ordered=True)
monthly_loans=monthly_loans.sort_values(['month'])
monthly_loans['period']=monthly_loans['month'].astype(str)+' '+monthly_loans['year'].astype(str)
plt.figure(figsize=(12,6))
plt.plot(monthly_loans['period'],monthly_loans['id'])
plt.fill_between(monthly_loans['period'], monthly_loans['id'], alpha=0.3)
plt.plot(monthly_loans['period'], monthly_loans['id'], marker='o')
for i in range(len(monthly_loans['id'])):
    val = monthly_loans['id'].iloc[i]
    plt.text(i, val + (val * 0.02),str(val), ha='center', fontsize=9, fontweight='bold', color='black')
plt.title('Total Loan application')    
plt.xlabel('Month', fontsize=12)
plt.ylabel('Total Loan Applications', fontsize=9)
plt.xticks(rotation=45)
plt.tight_layout()
plt.grid(True, alpha=0.3)
plt.show()
```

# TOTAL FUNDED AMOUNT

```python
#extact year and monthname
df['year']=df['issue_date'].dt.year
df['month']=df['issue_date'].dt.month_name()
#total loan application by year and month
monthly_loans=df.groupby(['year','month'])['loan_amount'].sum().div(1000000).reset_index()
month_order = ['January','February','March','April','May','June','July','August','September','October','November','December']
monthly_loans['month']=pd.Categorical(monthly_loans['month'],categories=month_order,ordered=True)
monthly_loans=monthly_loans.sort_values(['month'])
monthly_loans['period']=monthly_loans['month'].astype(str)+' '+monthly_loans['year'].astype(str)
plt.figure(figsize=(12,6))
plt.plot(monthly_loans['period'],monthly_loans['loan_amount'])
plt.fill_between(monthly_loans['period'], monthly_loans['loan_amount'], alpha=0.3)
plt.plot(monthly_loans['period'], monthly_loans['loan_amount'], marker='o')
for i in range(len(monthly_loans['loan_amount'])):
    val = monthly_loans['loan_amount'].iloc[i]
    plt.text(i, val + (val * 0.02), f"{val:.2f}", ha='center', fontsize=9, fontweight='bold', color='black') 

plt.xlabel('Month', fontsize=12)
plt.ylabel('Total Funded Amount', fontsize=9)
plt.xticks(rotation=45)
plt.tight_layout()
plt.grid(True, alpha=0.3)
plt.show()
```

# TOTAL AMOUNT RECIVED

```python
#extact year and monthname
df['year']=df['issue_date'].dt.year
df['month']=df['issue_date'].dt.month_name()
#total loan application by year and month
monthly_loans=df.groupby(['year','month'])['total_payment'].sum().div(1000000).reset_index()
month_order = ['January','February','March','April','May','June','July','August','September','October','November','December']
monthly_loans['month']=pd.Categorical(monthly_loans['month'],categories=month_order,ordered=True)
monthly_loans=monthly_loans.sort_values(['month'])
monthly_loans['period']=monthly_loans['month'].astype(str)+' '+monthly_loans['year'].astype(str)
plt.figure(figsize=(12,6))
plt.plot(monthly_loans['period'],monthly_loans['total_payment'])
plt.fill_between(monthly_loans['period'], monthly_loans['total_payment'], alpha=0.3)
plt.plot(monthly_loans['period'], monthly_loans['total_payment'], marker='o')
for i in range(len(monthly_loans['total_payment'])):
    val = monthly_loans['total_payment'].iloc[i]
    plt.text(i, val + (val * 0.02), f"{val:.2f}", ha='center', fontsize=9, fontweight='bold', color='black')
    
plt.xlabel('Month', fontsize=12)
plt.ylabel('Total Amount Recived', fontsize=9)
plt.xticks(rotation=45)
plt.tight_layout()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
donut_chart=df.groupby('term')['id'].count()
labels=df['term'].iloc[2:4].values
# colors
colors = ['#FF0000', '#0000FF']
# explosion
explode = (0.05, 0.05)

# Pie Chart
plt.pie(donut_chart, colors=colors,labels=labels, 
        autopct='%1.1f%%', pctdistance=0.85,
        explode=explode)

# draw circle
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
fig = plt.gcf()

# Adding Circle in Pie chart
fig.gca().add_artist(centre_circle)

# Adding Title of chart
plt.title('TOTAL LOAN APLLICATION')

# Displaying Chart
plt.show()
```

# HOME OWNER SHIP

```python
import matplotlib.pyplot as plt
import squarify  # for treemap visualization
# Get value counts
home_counts = df.groupby('home_ownership')['id'].count().sort_values(ascending=True)

# Create treemap
plt.figure(figsize=(8, 6))
squarify.plot(
    sizes=home_counts.values,
    label=[f"{label}\n{count}" for label, count in zip(home_counts.index, home_counts.values)],
    color=plt.cm.Paired.colors,  # you can change the color map
    alpha=0.8
)

plt.title("Home Ownership Distribution (Treemap)", fontsize=14, fontweight='bold')
plt.axis("off")  # hide axes
plt.show()
```
